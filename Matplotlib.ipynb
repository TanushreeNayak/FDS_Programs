```python
import numpy as np
import matplotlib.pyplot as plt

file = open('/content/drive/MyDrive/1 2.txt')
file.readline()

test1  = []
test2 = []

for row in file:
  row = row.split("\t")
  test1.append(int(row[0]))
  test2.append(int(row[1].strip("\n")))
x = np.array(test1)
y = np.array(test2)


plt.plot(x,y)

print(test1)
print(test2)
file.close()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.array([10,20,30])
y=np.array([40,10,30])
y1=np.array([20,40,10])

plt.plot(x,y,linestyle='dotted',color='purple')
plt.plot(x,y1,linestyle='-',color='red')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Plot with two or more lines with different styles')
plt.show()
```
